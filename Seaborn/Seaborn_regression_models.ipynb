```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sns.set_theme(color_codes=True)
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.regplot(data=tips, x='total_bill', y='tip')
```

```python
sns.lmplot(data=tips, x='total_bill', y='tip')
```

```python
sns.lmplot(data=tips, x='size', y='tip')
```

```python
import numpy as np

sns.lmplot(data=tips, x='size', y='tip', x_estimator=np.mean)
```

```python
anscombe = sns.load_dataset("anscombe")
```

```python
anscombe.head()
```

```python
anscombe['dataset'].value_counts()
```

```python
sns.lmplot(data=anscombe.query("dataset=='I'"), x='x', y='y', ci=None)
```

```python
sns.lmplot(data=anscombe.query("dataset=='II'"), x='x', y='y', ci=None)
```

```python
sns.lmplot(data=anscombe.query("dataset=='II'"), x='x', y='y', ci=None, order=2)
```

```python
sns.lmplot(data=anscombe.query("dataset=='III'"), x='x', y='y', ci=None)
```

```python
tips['big_tips'] = (tips.tip/tips.total_bill) > 0.2
```

```python
tips.head()
```

```python
sns.lmplot(data=tips, x='total_bill', y='big_tips')
```

```python
sns.lmplot(x="total_bill", y="tip", hue="smoker", data=tips)
```

```python
sns.lmplot(x="total_bill", y="tip", hue="smoker",
           col="time", row="sex", data=tips)
```

```python
sns.jointplot(data=tips, x='total_bill', y='tip', kind='reg')
```
