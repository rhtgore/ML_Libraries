```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
penguins = sns.load_dataset('penguins')
```

```python
penguins.head()
```

```python
#univariate histograms

sns.displot(data=penguins, x='bill_length_mm')
```

```python
sns.displot(data=penguins, x='bill_depth_mm')
```

```python
sns.displot(data=penguins, x='flipper_length_mm')
```

```python
sns.displot(data=penguins, x='flipper_length_mm',binwidth=3)
```

```python
sns.displot(data=penguins, x='flipper_length_mm',bins=20)
```
