```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sns.set_theme(style='darkgrid')
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.relplot(data=tips, x='total_bill', y='tip')
```

```python
sns.relplot(data=tips, x='total_bill', y='tip', hue='smoker')
```

```python
sns.relplot(data=tips, x='total_bill', y='tip', hue='smoker', style='smoker')
```

```python
sns.relplot(data=tips, x='total_bill', y='tip', hue='smoker', style='time')
```

```python
sns.relplot(data=tips, x='total_bill', y='tip', hue='size')
```

```python
sns.relplot(data=tips, x='total_bill', y='tip', size='size')
```

```python
sns.relplot(x="total_bill", y="tip", size="size", sizes=(15, 200), data=tips)
```

```python
import pandas as pd
import numpy as np

df = pd.DataFrame(dict(time=np.arange(500),
                       value=np.random.randn(500).cumsum()))
```

```python
df.head()
```

```python
sns.relplot(data=df, x='time', y='value', kind='line')
```

```python
df = pd.DataFrame(np.random.randn(500,2).cumsum(axis=0), columns=["x", "y"])
```

```python
df.head()
```

```python
sns.relplot(data=df, x='x', y='y', kind='line', sort=False)
```

```python
fmri = sns.load_dataset('fmri')
```

```python
fmri.head()
```

```python
sns.relplot(data=fmri, x='timepoint', y='signal', kind='line')
```

```python
sns.relplot(data=fmri, x='timepoint', y='signal', kind='line', ci=None)
```

```python
sns.relplot(data=fmri, x='timepoint', y='signal', kind='line', ci='sd')
```

```python
sns.relplot(data=fmri, x='timepoint', y='signal', kind='line', estimator=None)
```

```python
sns.relplot(data=fmri, x='timepoint', y='signal', hue='event', kind='line')
```

```python
sns.relplot(data=fmri, x='timepoint', y='signal', hue='region', style='event', kind='line')
```

```python
sns.relplot(data=fmri, x='timepoint', y='signal', hue='region', style='event', dashes=False, markers=True, kind='line')
```
